## Check Missing Values

```python
df.isna().sum()
```

```python
df_c = df.fillna('Unknown')
```

```python
df.columns
```

```python
categoricals=['Job', 'Marital', 'Education', 'Default', 'Communication', 'LastContactDay', 'LastContactMonth', 'Outcome']
```

```python
for i in categoricals:
    print(f"{i} distribution: \n {df_c[i].value_counts()}")
    plt.hist(df_c[df_c['CarInsurance']==0][i], color='blue', alpha=0.5, label ='No')
    plt.hist(df_c[df_c['CarInsurance']==1][i], color='green', alpha=0.5, label ='Yes')
    plt.title(i + " Distribution")
    plt.show()
    print('')
# print(f"Default distribution: \n {df.Default.value_counts()}")
```
