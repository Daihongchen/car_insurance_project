```python
def factorial(n):
    if n==1:
        return 1
    else: 
        return n*factorial(n-1)
```

```python
factorial(10)/factorial(8)
```

```python
results=[]
for i in range(1,10,1):
    results.append(factorial(10)/factorial(10-i)/factorial(i))
    
```
